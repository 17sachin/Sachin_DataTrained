```python
#Check the Correlation

df_cor=df.corr()
df_cor
```

```python
#For better understanding we can see it via visualization (with the help of heatmap)
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(df_cor, annot=True , cmap='winter')
```

```python
#Obseravtion:
    #class is negatively correlated with ra,dec,g,run,camcol,redshift
    #class has positive correlation with plate,mjid,fiberid,specobjid
    
```

```python
plt.figure(figsize=(8,6))
df.drop('class',axis=1).corrwith(df['class']).plot(kind='bar',grid=True)
plt.xticks(rotation=20)
plt.title("Correlation with target Variable")
```

```python
#Observation:
    #ra,dec,u,g,run,redshift is neagtively correlated with the class
    #r,i,z,specobjid,plate,mjid ,fiberid is postively correlated with class
```
